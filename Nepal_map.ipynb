```python
df = pd.read_csv('cleandata.csv')
```

```python
coor = pd.DataFrame(df['Geographical Region'])
coor
```

```python
coor = coor['Geographical Region'].str.split(',',expand = True)

coor.columns = ['Latitude', 'Longitude']
```

```python
coor
```

```python
coor['Latitude'] = coor['Latitude'].astype('float')
coor['Longitude'] = coor['Longitude'].astype('float')
```

```python
coor.plot('Longitude','Latitude',kind = 'scatter')
```

```python
fig, ax = plt.subplots(figsize=(8,6))
# plot map on axis
countries = gpd.read_file(  
     gpd.datasets.get_path("naturalearth_lowres"))
countries[countries["name"] == "Nepal"].plot(color="lightgrey",
                                                 ax=ax)
```

```python
fig, ax = plt.subplots(figsize=(8,6))
# plot map on axis
countries = gpd.read_file(gpd.datasets.get_path("naturalearth_lowres"))
countries[countries["name"] == "Nepal"].plot(color="lightgrey",ax=ax)
coor.plot('Longitude','Latitude',kind = 'scatter',ax=ax, c = 'red')
ax.grid(b=True, alpha=0.5)
plt.show()
```

```python
fig = px.scatter_geo(coor, lat='Latitude', lon='Longitude', color='Latitude',title='Incidents in Nepal')
fig.show()
```
